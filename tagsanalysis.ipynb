```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tags = pd.read_csv('D:/desktop/keywords.csv')
```

```python
tags = tags[['_id','count','idf']]
sub = tags[:1000]
sub['idf'] = np.log10(sub['idf'])
avg = np.mean(sub['idf'])
std = np.std(sub['idf'])
max = np.max(sub['idf'])
min = np.min(sub['idf'])
quart3 = np.quantile(sub['idf'],.75)
quart1 = np.quantile(sub['idf'],.25)
IQR = quart3-quart1
top = avg +3*std
bottom = avg +-3*std


print(avg,std,max,min)
```

```python
fig, ax = plt.subplots()
n,bins,patches = ax.hist(sub['idf'],100)
plt.axvline(x=top)
plt.axvline(x=bottom)

plt.show()
```
